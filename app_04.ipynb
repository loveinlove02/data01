## ③ 꺽은선 그래프 그리기

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic' 

data = {
            '이름':['김철수', '이영희', '김민수'],    
            '국어':[90, 70, 80],
            '영어':[99, 77, 88],
            '수학':[75, 85, 90]
       }

df = pd.DataFrame(data)
df = df.set_index('이름')
print(df)

name = df.index.values

fig, ax = plt.subplots()
ax.plot(name, df['국어'], color='blue', label='국어')
ax.plot(name, df['영어'], color='#FE7D27', label='영어')
ax.plot(name, df['수학'], color='green', label='수학')
ax.legend(loc='upper right')
plt.title('성적 그래프')
plt.show()
```

## ④ 막대 그래프 그리기

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['axes.unicode_minus'] = False    

data = {
            '이름':['김철수', '이영희', '김민수'],    
            '국어':[90, 70, 80],
            '영어':[99, 77, 88],
            '수학':[75, 85, 90]
       }

df = pd.DataFrame(data)
df = df.set_index('이름')
# print(df)

N = df.shape[0]
index = np.arange(N)  # [0 1 2]
name = df.index.values

fig, ax = plt.subplots()

w = 0.25

ax.bar(index - w, df['국어'], color='blue', label='국어', width=0.25)
ax.bar(index, df['영어'], color='red', label='영어', width=0.25)
ax.bar(index + w, df['수학'], color='green', label='수학', width=0.25)

ax.set_xticks(index, name)
ax.tick_params(axis='x', rotation=40)
ax.tick_params(axis='y', rotation=20)
ax.legend(loc='upper right', ncol=3)

plt.show()
```

## ⑤ 원 그래프 그리기

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['axes.unicode_minus'] = False    

data = {
            '이름':['김철수', '이영희', '김민수'],    
            '국어':[90, 70, 80],
            '영어':[99, 77, 88],
            '수학':[75, 85, 90]
       }

df = pd.DataFrame(data)
df = df.set_index('이름')
# print(df)
# print(df['국어'])
# print(df['국어'].iloc[0])

fig, ax = plt.subplots()

values = [df['국어'].iloc[0], df['국어'].iloc[1], df['국어'].iloc[2]]
labels = df.index.values
# labels = [df.index[0], df.index[1], df.index[2]]
explode = [0.02] * 3
colors = ['#ffadad', '#ffd6a5', '#a0c4ff']

ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode, colors=colors)
# wedgeprops = {'width': 0.6, 'edgecolor':'white', 'linewidth':5}
# ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode, colors=colors, wedgeprops=wedgeprops)
ax.legend(loc=(1.0, 0.3), title='국어 성적')

plt.title('성적 그래프')
plt.show()
```

## ⑥ 과목별 원 그래프 그리기

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['axes.unicode_minus'] = False    

data = {
            '이름':['김철수', '이영희', '김민수'],    
            '국어':[90, 70, 80],
            '영어':[99, 77, 88],
            '수학':[75, 85, 90]
       }

df = pd.DataFrame(data)
df = df.set_index('이름')

print(df)

values1 = df['국어'].values
values2 = df['영어'].values
values3 = df['수학'].values
labels = df.index.values

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].pie(values1, labels=labels, autopct='%1.1f%%', startangle=90)
axes[0].set_title('국어점수 비율')

axes[1].pie(values2, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1].set_title('영어점수 비율')

axes[2].pie(values3, labels=labels, autopct='%1.1f%%', startangle=90)
axes[2].set_title('수학점수 비율')

plt.suptitle('성적 그래프')
plt.show()
```

## ⑦ 과목별 막대그래프 그리기

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['axes.unicode_minus'] = False    

data = {
            '이름':['김철수', '이영희', '김민수'],    
            '국어':[90, 70, 80],
            '영어':[99, 77, 88],
            '수학':[75, 85, 90]
       }

df = pd.DataFrame(data)
df = df.set_index('이름')
# print(df)

values1 = df['국어'].values
values2 = df['영어'].values
values3 = df['수학'].values
names = df.index.values

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].bar(names, values1, width=0.6)
axes[0].set_xlabel('이름')
axes[0].set_title('국어 점수')

axes[1].bar(names, values2, width=0.6)
axes[1].set_xlabel('이름')
axes[1].set_title('영어 점수')

axes[2].bar(names, values3, width=0.6)
axes[2].set_xlabel('이름')
axes[2].set_title('수학 점수')

plt.suptitle('성적 그래프')             
plt.show()
```
