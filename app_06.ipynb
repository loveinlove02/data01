## 1. csv 파일 불러오기
### 불러온 파일을 데이터프레임으로 만든다.

```python
df = pd.read_csv('인구동태건수_및_동태율_추이_출생_사망_혼인_이혼__20240710153342.csv', encoding='cp949')
df
```

```python
df.iloc[53, 0] = df.iloc[53, 0][:4]
df
```

### 새로운 데이터프레임을 만든다. 5년 단위로 가져온다.

```python
new_df = pd.DataFrame(
    {
        '년도': df['시점'][::5],				
        '출생아수': df['출생아수(명)'][::5],
        '합계출산율': df['합계출산율(명)'][::5]
    }
)
```

```python
new_df
```
