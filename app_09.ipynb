```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False    
```

## 우리나라 남녀 인구피라미드

### 1. csv파일을 읽어온다

```python
df = pd.read_csv('성_및_연령별_추계인구_1세별__5세별____전국_20240712133223.csv', encoding='cp949')
df
```

### 2. 필요없는 행을 삭제

```python
df.drop(labels=[0, 1, 2], axis=0, inplace=True)
```

```python
df
```

### 3. 인덱스를 재지정

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

### 4. 컬럼 이름을 바꾼다

```python
df.rename(columns={'2024':'전체', '2024.1':'남자', '2024.2':'여자'}, inplace=True)
```

```python
df
```

### 5. 연령을 다시 지정한다

0-9 <br>
10-19 <br>
20-29 <br>
30-39 <br> 
40-49 <br>
50-59 <br>
60-69 <br>
70-79 <br>
80-

```python
for i in range(0, len(df), 2):
    print(i, i+1)
```

```python
print(df['연령별'].iloc[0], df['전체'].iloc[0], df['남자'].iloc[0], df['여자'].iloc[0])
print(df['연령별'].iloc[1], df['전체'].iloc[1], df['남자'].iloc[1], df['여자'].iloc[1])
a = int(df['전체'].iloc[0])+int(df['전체'].iloc[1])
b = int(df['남자'].iloc[0])+int(df['남자'].iloc[1])
c = int(df['여자'].iloc[0])+int(df['여자'].iloc[1])
print(f'전체: {a}')
print(f'남자: {b}')
print(f'여자: {c}')
```

```python
for i in range(0, len(df), 2):
    print(df['연령별'].iloc[i].replace('세',''), 
          df['연령별'].iloc[i+1].replace('세',''), 
          int(df['남자'].iloc[i]), int( df['남자'].iloc[i+1]),
		  int(df['남자'].iloc[i])+int( df['남자'].iloc[i+1]))
```

```python
man_population = []  # 남자 인구수

for i in range(0, len(df), 2):
    man_population.append(int(df['남자'].iloc[i])+int( df['남자'].iloc[i+1]))
    
man_population
```

```python
woman_population = []  # 여자 인구수

for i in range(0, len(df), 2):
    woman_population.append(int(df['여자'].iloc[i])+int( df['여자'].iloc[i+1]))

woman_population
```

```python
age_group = []  # 연령 컬럼 다시 만들기

for i in range(0, 10, 1):
    print(str(i*10) + '-' + str((i*10)+9))
    age_group.append(str(i*10) + '-' + str((i*10)+9))

age_group    
```

### 6. 데이터프레임을 새로 만든다

```python
new_df = pd.DataFrame(
    {
        '연령대': age_group,
        '남자': man_population,
        '여자': woman_population
	}
)

new_df
```

```python
age_group2 = age_group[0:8] + ['80-']
age_group2
```

```python
man_population2 = man_population[0:8] + [sum(man_population[8:])]
man_population2
```

```python
woman_population2 = woman_population[0:8] + [sum(woman_population[8:])]
woman_population2
```

```python
new_df2 = pd.DataFrame(
    {
        '연령대': age_group2, 
        '남자': man_population2, 
        '여자': woman_population2
	}
)

new_df2
```

### 7. 막대그래프 그리기

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.barh(new_df2['연령대'], new_df2['여자']//10000)
ax.barh(new_df2['연령대'], -new_df2['남자']//10000)
ax.set_xticks(np.arange(-500, 600, 100))	# 수정

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.barh(new_df2['연령대'], new_df2['여자']//10000)
ax.barh(new_df2['연령대'], -new_df2['남자']//10000)

for i in range(0, len(new_df2), 1):
    ax.text(int(new_df2.iloc[i]['여자']//10000)+30, new_df2.iloc[i]['연령대'], 
			int(new_df2.iloc[i]['여자']//10000), 
			fontsize=12,
			color='gray', 
			fontweight='bold', 
			va='center',
			ha='center',
			bbox=dict(boxstyle='round', facecolor='white', alpha=0.4, edgecolor="blue", pad=0.2) 
        	)
	

for i in range(0, len(new_df2), 1):
    ax.text(int(-new_df2.iloc[i]['남자']//10000) - 30, new_df2.iloc[i]['연령대'], 
			int(new_df2.iloc[i]['남자']//10000), 
			fontsize=12,
			color='gray', 
			fontweight='bold', 
			va='center',
			ha='center',
			bbox=dict(boxstyle='round', facecolor='white', alpha=0.4, edgecolor="#5C9DFF", pad=0.2) 
        	)
    
    
ax.set_xticks(np.arange(-500, 600, 100))	# 수정

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.barh(new_df2['연령대'], new_df2['여자']//10000, label='여자')
ax.barh(new_df2['연령대'], -new_df2['남자']//10000, label='남자')

for i in range(0, len(new_df2), 1):
    ax.text(int(new_df2.iloc[i]['여자']//10000)+30, new_df2.iloc[i]['연령대'], 
			int(new_df2.iloc[i]['여자']//10000), 
			fontsize=12,
			color='gray', 
			fontweight='bold', 
			va='center',
			ha='center',
			bbox=dict(boxstyle='round', facecolor='white', alpha=0.4, edgecolor="blue", pad=0.2) 
        	)
	

for i in range(0, len(new_df2), 1):
    ax.text(int(-new_df2.iloc[i]['남자']//10000) - 30, new_df2.iloc[i]['연령대'], 
			int(new_df2.iloc[i]['남자']//10000), 
			fontsize=12,
			color='gray', 
			fontweight='bold', 
			va='center',
			ha='center',
			bbox=dict(boxstyle='round', facecolor='white', alpha=0.4, edgecolor="#5C9DFF", pad=0.2) 
        	)
    
    
ax.set_xticks(np.arange(-500, 600, 100))	

ax.set_xlabel('인구(만명)', labelpad=10, loc='center', color='#7777A0', fontdict={'fontsize': 18, 'fontweight':'bold'})        
ax.set_ylabel('연령대', labelpad=20, loc='center', color='#8EB8F0', fontdict={'fontsize': 18, 'fontweight':'bold'})

ax.tick_params(axis='y',  direction='in', length=9, pad=10, labelsize=15, labelcolor='blue', width=2, color='red')
ax.tick_params(axis='x',  direction='inout', length=9, pad=10, labelsize=15, labelcolor='gray', width=2, color='red', labelrotation=20)

ax.legend(loc=(-0.1, -0.15), fontsize=12, shadow=True, ncol=2)
plt.grid(True, axis='x', color='gray', alpha=0.2, linestyle='-', linewidth=1)
plt.show()
```
