## 1.데이터프레임

데이터프레임은 데이터를 다룰때 가장 많이 사용하는 데이터 형태로, 행과 열로 구분된 사각형 표처럼 생겼습니다.

### ‘열’ 은 속성이다.

세로로 나열되는 열은 속성을 나타냅니다. 열은 컬럼(column)이라고 부릅니니다.



### ‘행’ 은 한 사람의 정보이다

가로로 나열되는 행은 각 사람의 정보를 나타냅니다. 행은 로우(row) 라고 부릅니니다.



만약 어떤 데이터에 10명의 정보가 들어있다면, 그 데이터에는 10개의 row를 가지고 있다고
표현할 수 있습니다.

## 1.1 데이터프레임 만들기

데이터 프레임을 생성할 때는 data, index, columns를 인자로 넣어서 만듭니다.<br>
리스트를 data에 넣고, index, columns에 행(row)과 열(column) 레이블을 지정해서 생성합니다.

### 데이터프레임 만들기

```python
import pandas as pd

df = pd.DataFrame([[1,2,3,4], 
                   [5,6,7,8]], index=['a', 'b'], columns=['col1','col2','col3','col4'])

df
```

### 데이터프레임 속성확인

```python
import pandas as pd

df = pd.DataFrame([[1,2,3,4], 
                   [5,6,7,8]], index=['a', 'b'], columns=['col1','col2','col3','col4'])

print(df.shape)      # 모양
print(df.ndim)       # 차원
print(df.size)       # 크기
print(df.index)      # index의 속성
print(df.columns)    # columns의 속성
```

### 시리즈를 딕셔너리에 넣어서 데이터프레임 만들기

```python
import pandas as pd

sr1 = pd.Series([1, 2, 3, 4], 		name='sr1')
sr2 = pd.Series(['a','b','c','d'], 	name='sr2')

# 생성된 시리즈를 가지고 딕셔너리에 넣어서 데이터 프레임을 만든다
# 딕셔너리의 키 s1, s2는 데이터 프레임의 열의 이름이 된다. 
df = pd.DataFrame({'s1': sr1, 's2': sr2})

df
```
